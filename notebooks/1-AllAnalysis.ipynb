# Load data

```python
# Path to where the image is stored
path = "/home/u2175049/Downloads/MarioMovie/"

# Name of the segmented image
seg_filename = "220801_Fish_2_straight_Z_stack_left_labels.tif"

# Update pixel size to have the correct measure
pixel_size = dict(x_size = 0.21,
                  y_size = 0.21,
                  z_size = 0.5,)

```

```python
# Read the image
image = io.imread(os.path.join(path, seg_filename))
np.unique(image)
```
