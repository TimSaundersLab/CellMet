# Import

```python
import os
import numpy as np
import pandas as pd

from skimage import io

import matplotlib.pyplot as plt

from CellSeg.segmentation import Segmentation
from CellSeg.analysis import cell_analysis
```

# Load data

```python
# Path to where the image is stored
path = "/mnt/sda1/CellSegPaperDatas/Artificial_image/img_size/"

# Name of the segmented image
seg_filename = "artificial_tissue_70cells_250px_5.tif"

# Update pixel size to have the correct measure
pixel_size = dict(x_size = 0.21,
                  y_size = 0.21,
                  z_size = 0.5,)

```

```python
# Read the image
image = io.imread(os.path.join(path, seg_filename))
```

```python
# Show the middle slice of the image
fig, ax = plt.subplots()
ax.imshow(image[int(image.shape[0]/2)], cmap="jet")
```

# Analyse cell

```python
# Create the segmentation object
seg = Segmentation(image, pixel_size, path)
```

```python
# Perform the prerequisite for faster analysis
seg.perform_prerequisite(save_mesh=False, overwrite=True)
```

```python
# Analyse cell 
cell_analysis(seg, parallelized=True, degree_convert=True)
```
